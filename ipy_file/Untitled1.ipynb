```python
path = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
```

```python
path = os.path.join(path, 'output')
```

```python
path
```

```python
data = {}
for f in os.listdir(path):
    if 'cost.npy' not in f: continue
    name = f[0:3]
    print('loading ' + name)
    data[name] = {}
    data[name]['cost'] = np.load(path + '/' + f)
    data[name]['acc']= np.load(path + '/' + name + '_acc.npy')
```

```python
label = []
plt.figure(figsize=(15,10))
for k in data.keys():
    label.append(k)
    plt.plot(data[k]['acc'][100:], data[k]['cost'][100:],'o')
plt.legend(label, loc='upper right', prop={'size':15})
plt.ylabel('Sum of difference')
plt.xlabel('loss')
plt.show()
```

```python
label = []
plt.figure(figsize=(15,10))
for k in data.keys():
    label.append(k)
    plt.plot(range(len((data[k]['acc'][100:]))),data[k]['acc'][100:],'--')
plt.legend(label, loc='upper right', prop={'size':15})
plt.ylabel('Sum of difference')
plt.xlabel('loss')
plt.show()
```
